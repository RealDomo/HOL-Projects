```python
import pandas as pd

ps = {
      'Song': ['Money Trees', 'Love Yours', 'Trust Issues', 'Swimming Pools', 'wacced out murals', 'Ultralight Beam', '30 for 30', 'Stan'],
      'Artist': ['Kendrick Lamar', 'J. Cole', 'Drake', 'Kendrick Lamar', 'Kendrick Lamar', 'Kanye West', 'SZA', 'Eminem'],
      'Energy': [0.9, 0.4, 0.5, 0.6, 0.9, 0.5, 0.5, 0.6],
      'Day_Played': ['Saturday', 'Friday', 'Saturday', 'Sunday', 'Tuesday', 'Monday', 'Thursday', 'Wednesday']
      }

playlist = pd.DataFrame(ps)

print(playlist, "\n")

print(playlist.head())
```

```python
artistCounts = playlist['Artist'].value_counts()

print(artistCounts.index[0])
```

```python
averageEnergy = playlist['Energy'].mean()

if averageEnergy < 0.4:
  print("I'm just a chill guy")
elif averageEnergy > 0.6:
  print("I like high energy music")
else:
  print("I like medium energy music")

print(averageEnergy)
```

```python
import matplotlib.pyplot as plt

day_counts = playlist['Day_Played'].value_counts()
plt.bar(day_counts.index, day_counts.values)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.style.use('ggplot')
plt.show()

print("I think I listen to music mostly on Saturday because I don't have school.")
```

```python
import matplotlib.pyplot as plt

plt.bar(artistCounts.index, artistCounts.values)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.style.use('dark_background')
plt.show()

print("I think this is meaningful because it shows the number of songs by each artist graphically.")
```
